### *Matrix operations*

```python
M = A.reshape(4,5)
```

if you want to find the transpose of this matrix, you can simply say M dot capital
T, this is a transpose of this matrix.

```python
M.T
```

```python
M*M
```

```python
N = M.transpose()
N
```

if I
say M star N, this will not work; because M star, star means term by term multiplication.
If you want to use matrix multiplication; actually matrix multiplication of two matrices
can be thought of as extension of dot product.

```python
M.dot(N) # matrix multipication
```

```python
M@N # Another way of matrix multiplication 
```

```python
D = M@N
D
```

```python
# for determinant
Determinat = la.det(D)
Inverse = la.inv(D)
print(f'The determinant and inverse of matrix D is {Determinat} and {Inverse}')
```

```python
la.matrix_rank(D)
```

```python
D.trace()
```

```python
import numpy as np

# Define a 4x4 matrix
A = np.array([[4, -2, 1, 3],
              [1, 3, -2, 4],
              [2, -1, 3, 5],
              [1, 2, 3, 4]])

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

# Print the eigenvalues
print("Eigenvalues of the matrix:")
print(eigenvalues)

# Print the eigenvectors
print("\nEigenvectors of the matrix:")
print(eigenvectors)
```

```python
import numpy as np

# Define a 3x3 matrix
A = np.array([[1, 1, 0],
              [1, 1, 0],
              [0, 0, 0]])

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

# Print the eigenvalues
print("Eigenvalues of the matrix:")
print(eigenvalues)

# Print the eigenvectors
print("\nEigenvectors of the matrix:")
print(eigenvectors)
```

### **Python Graphics using MatplotLib**

So, first of all, let us see how does one plot graphs in Python. Plotting graphs in Python is done through a module known as matplotlib, which is a comprehensive library for creating static interactive and animated graphics in Python. One can use pyplot and pylab modules inside matplotlib to plot graphs of various objects inside Python. Pylab actually is a combination of pyplot functionality and NumPy functionality in a single namespace. We will be using pyplot mostly.

Next, in order to plot set of points, let us generate random 200 points, which are
normally distributed with mean 10 and standard deviation 5. So both x and y are 200
random points which are normally distributed with mean 10 and standard deviation 5.

#### *Scatter Plots*

```python
pip install matplotlib
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.normal(10,5,200)
y = np.random.normal(10,5,200)
plt.scatter(x,y)
plt.show()
```

### Changing point colour, size, adding labels for axes and title to the plot


```python
plt.scatter(x, y, c='r', s=100)
plt.title("Scatter Plot")
plt.xlabel('$x$-axis')
plt.ylabel('$y$-axis')
plt.grid(False)
plt.show()
```

### *Plotting graph of explicitely defined function*

plot the graph of function $ f(x) = e^x $ 

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-5,5,100)
y = np.exp(x)
plt.plot(x,y)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the range for x-values
x = np.linspace(-5, 5, 400)

# Calculate the corresponding y-values using the exponential function
y = np.exp(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='$f(x) = e^x$')

# Add labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of f(x) = e^x')

# Add a legend
plt.legend()

# Add a grid for better readability
plt.grid(True)

# Show the plot
plt.show()
```
