*SymPy is an excellent open-source Python library for symbolic computation. Its **dsolve()** function is particularly useful for finding analytical solutions to ordinary differential equations (ODEs) and systems of ODEs. This tool provides a simple workflow for solving both single-equation and coupled systems, as shown in the examples that follow.*

```python
import sympy as sp
from sympy import symbols, Function, Eq, dsolve, Derivative, sqrt
```

```python
from sympy import *
# Enable pretty printing
init_printing(use_unicode=True)
```

*Here, symbols defines independent variables, Function represents unknown functions, Eq constructs equations, dsolve solves them, and Derivative builds derivative objects when needed.*

*Consider the classic linear ODE:*

 $$ \frac{dy}{dx} +y(x)=e^{x} $$

```python
# Define symbols and function
x = symbols('x')
y = Function('y')

# Define the differential equation
ode = Eq(Derivative(y(x), x) + y(x), sp.exp(x))

# Solve the ODE
solution = dsolve(ode)
solution
```

```python
x = symbols('x')
y = Function('y')
ode = Eq(Derivative(y(x), x), (x**2+2)/y(x))
solution = dsolve(ode,y(x))
solution
```

```python
ode1 = Eq(Derivative(y(x),x)+x*y(x),x**2*y(x))
sol = dsolve(ode1)
sol
```

```python
## Second Order D.E.
ode2 = Eq(Derivative(y(x), x, 2) + y(x),0)
sol = dsolve(ode2)
sol
```

```python
ode2 = Eq(Derivative(y(x), x, 2) + y(x), sin(x))
sol = dsolve(ode2)
sol
```

```python
ode3 = Eq(Derivative(y(x), x, 3) + y(x), exp(x))
sol = dsolve(ode2)
sol
```

```python
# Handling initial value problem
x = symbols('x')
y = Function('y')
ode = Eq(Derivative(y(x), x), (x**2+2)/y(x))
ini = ics = {y(0): 1} # y(0) = 1
solution = dsolve(ode,y(x), ics = ini)
solution
```

```python
# Define variable and function
x = symbols('x')
y = Function('y')

# Define the differential equation: y'' + y = 0
ode = Eq(Derivative(y(x), x, 2) + y(x), 0)

# Solve the ODE with initial conditions y(0) = 1, y'(0) = 0
solution = dsolve(ode, ics={y(0): 1, y(x).diff(x).subs(x, 0): 0})
solution
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define variable and function
x = sp.symbols('x')
y = sp.Function('y')

# Define the differential equation: y'' + y = 0
ode = sp.Eq(sp.Derivative(y(x), x, 2) + y(x), 0)

# Solve with initial conditions y(0)=1, y'(0)=0
sol = sp.dsolve(ode, ics={y(0): 1, y(x).diff(x).subs(x, 0): 0})
print("Solution:", sol)

# Extract the RHS (y(x) expression)
y_sol = sol.rhs

# Convert symbolic solution to a numerical function
y_func = sp.lambdify(x, y_sol, "numpy")

# Generate x values
x_vals = np.linspace(0, 10, 400)
y_vals = y_func(x_vals)

# Plot
plt.figure(figsize=(8,5))
plt.plot(x_vals, y_vals, label=f"$y(x) = {sp.latex(y_sol)}$")
plt.title("Solution of $y'' + y = 0$ with $y(0)=1, y'(0)=0$")
plt.xlabel("x")
plt.ylabel("y(x)")
plt.grid(True)
plt.legend()
plt.show()
```

#### *Solving a system of ODEs*

```python
from sympy import *
# Enable pretty printing
init_printing(use_unicode=True)

t = symbols('t')
x = Function('x')
y = Function('y')

eq1 = Eq(Derivative(x(t), t),  3*x(t) - 4*y(t))
eq2 = Eq(Derivative(y(t), t),  4*x(t) + 3*y(t))

sol_system = dsolve((eq1, eq2))
sol_system
```

```python
sol = dsolve((eq1, eq2), ics = {x(0): 1, y(0): 1})
sol
```
